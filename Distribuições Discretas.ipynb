Uma variável aleatória é tida como uma que apresenta um valor único para cada elemento, sendo esse valor determinado aleatóriamente. Podendo ser contínuas e discretas.



### Distribuição de Bernoulli

A distribuição de Bernoulli é simples e se resume a sucesso e fracasso, isto é, à variáveis dicotómicas. \\
Onde $p=$ probabilidade de sucesso e $q=(1-p)$ é a probabilidade de fracasso. \\
Logo, essa distribuição fornece a probabilidade de sucesso ou fracasso de uma variável aleatória $X$ em um **único experimento**. A denotação para uma variável que segue a distribuição de bernoulli é: $X$~Bern$(p)$. \\
Sua FDP(PMF) é dada por $P(X=x)=P^{x} *(1-P)^{1-X}$ \\

$E(X)=p$ \\
$Var(X)=p*(1-p)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
p=0.6 #se P(X=x)=p se k=1 e (1-p) se k=0
mean,var, skew, kurt=bernoulli.stats(p, moments='mvsk')
print(f"E(X)=",mean, "\n Var(X)=",var, "\n Assimetria=", skew, "\n Curtose=",kurt)

#Gráfico Função de Densidade acumulada (FDA)
fig, ax = plt.subplots(1, 1)
x = np.arange(0,1)
ax.plot(x, bernoulli.pmf(x, p), 'bo', ms=8, label='bernoulli pmf')
ax.vlines(x, 0, bernoulli.pmf(x, p), colors='b', lw=5, alpha=0.5)
rv = bernoulli(p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
        label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```

### Distribuição Binomial

Consiste um experimento de $n$ **repetições independentes de um experimento de Bernoulli** com probabilidade $p$ de **sucesso que se mantém constante em todas repetições**. \\

Uma variável $X$ que segue uma distribuição Binomial é representada por $X$~b$(n,p)$ e sua FDP é dada por: $\left(\begin{matrix}n\\k\\\end{matrix}\right)\ast p^k\ast\left(1-p\right)^{n-k},\ \ \ \ \ k=0,1,2,..,n$, em que $\left(\begin{matrix}n\\k\\\end{matrix}\right)$ é  $\frac{n!}{k!(n-k)!}$. \\

Sua média é dada por $E(X)=n*p$ \\
e variância por $Var(X)=n*p*(1-p)$

```python
from scipy.stats import binom
```

```python
p=0.99
n=30
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
print(f"E(X)=",mean, "\n Var(X)=",var, "\n Assimetria=", skew, "\n Curtose=",kurt)
```

#### Exercício Distribuição Binomial
Determinada peça é produzida em uma linha de produção.A probabilidade de que a peça não tenha defeitos é de 99%. Se forem produzidas 30 peças, qual a probabilidade de que pelo menos 28 delas esteja em boas condições? Determine também a média e a variância da variável aleatória. 

$X =$ variável aleatória que representa o número de sucessos (peças em boas condições) nas 30 repetições; \\
$p =$ 0,99 =probabilidade de que a peça esteja em boas condições; \\
$q =$ 0,01 =probabilidade de que a peça seja defeituosa; \\
$n =$ 30 repetições; \\
$k =$ número de sucessos. 



```python
p=0.99
n=30
probabilidade=binom.pmf(28,n,p)+binom.pmf(29,n,p)+binom.pmf(30,n,p)
print('a probabilidade é de:', '%0.4f' %probabilidade)

#Função de densidade acumulada
fig, ax = plt.subplots(1, 1)
x = np.arange(0,30)
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
        label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```

### Distribuição Geométrica

Assim como a Binomial considera sucessivos ensaios de Bernoulli independentes, com probabilidade de sucesso $p$. Contudo, ao invés de ser realizada por um número fixo de tentativas ela é realizada até que o primeiro sucesso seja obtido. Ela possui duas parametrizações distintas. \\

A primeira parametrização considera sucessivos ensaios de Bernoulli independentes, com probabilidade de sucesso $p$ em cada ensaio, até que ocorra um sucesso. e não inclui o zero como resultado possível. \\

A segunda parametrização conta o número de falhas ou fracassos antes do primeiro sucesso. Como aqui é possível obter sucesso já no primeiro ensaio de Bernoulli, incluímos o zero como resultado possível. \\

Seja $X$ a variável aleatória que representa o número de tentativas até o primeiro sucesso.A variável X tem distribuição geométrica com parâmetro $p$, denotada por $X$~Geo$(p)$, se sua função de probabilidade for dada por:
$P(X=x)=p\ast{(1-p)}^{(x-1)}$, $x=1,2,3...$

Para o segundo caso, consideremos Y a variável aleatória que representa o número de falhas ou fracassos antes do primeiro sucesso:
$P(Y=y)=p\ast{(1-p)}^y, \ \ y=0,1,2,3...$

Sua média e variância é dada por:
$E(X)=\frac{1}{p}$ \\
$Var(x)=\frac{(1-p)}{p²}$ \\

$E(Y)=\frac{(1-p)}{p}$ \\
$Var(Y)=\frac{(1-p)}{p²}$ \\

```python
from scipy.stats import geom
```

#### Exercício Distribuição Geométrica

Uma empresa fabrica determinado componente eletrônico, de modo que, ao final do processo, cada componente é testado, um a um. Suponha que a probabilidade de um componente eletrônico estar defeituoso seja
de 0,05. Determine a probabilidade de que o primeiro defeito seja encontrado no oitavo componente testado.
Calcule também o valor esperado e a variância da variável aleatória.

```python
probabilidade=geom.pmf(8,0.05)
p = 0.05
mean, var, skew, kurt = geom.stats(p, moments='mvsk')
print('a probabilidade é de:', '%0.4f' %probabilidade,
      "\n E(X)=",mean, "\n Var(X)=",var, "\n Assimetria=", skew, "\n Curtose=",kurt)

#FDA
fig, ax = plt.subplots(1, 1)
x = np.arange(geom.ppf(0.01, p),
              geom.ppf(0.99, p))
ax.plot(x, geom.pmf(x, p), 'bo', ms=8, label='geom pmf')
ax.vlines(x, 0, geom.pmf(x, p), colors='b', lw=5, alpha=0.5)
rv = geom(p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
        label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```

### Distribuição Binomial Negativa(distribuição de Pascal)

Realiza sucessivos ensáios de Bernoulli independenes com probabilidade de sucesso constante até atingir número prefixado de sucessos $(k)$. Seja $X$ a variável aleatória, ela possui distribuição binomial negativa, denotada por $X$~bn$(k,p)$ e sua FDP é dada por: \\
$\left(\begin{matrix}x-1\\k-1\\\end{matrix}\right)\ast p^k\ast\left(1-p\right)^{x-k},\ \ \ \ \ x=k,k+1,....$

Sua média e variância são denotadas por: \\
$E(X)= \frac{k}{p}$ \\
$Var(X)=\frac{k*(1-p)}{p²}$ \\

Diferente da binomial, na binomial negativa deve-se fixar o número de sucessos $(k)$ e observar o número de ensáios de Bernoulli necessários para sua obtenção. Enquanto na Binomial se fixa o número de ensáios de Bernoulli (amostra) e observa o númro de sucessos.


#### Exercício distribuição binomial negativa

Suponha que um aluno acerte três questões a cada cinco testes. Seja $X$ o número de tentativas até o décimo segundo acerto. Determine a probabilidade de que o aluno precise fazer 20 questões para acertar 12

```python
from scipy.stats import nbinom
p=0.6
k=8 #20 - 12
n=12 #Décimo segundo acerto
mean, var, skew, kurt = nbinom.stats(n, p, moments='mvsk')
probabilidade=nbinom.pmf(k,n,p)
print('a probabilidade é de:', '%0.4f' %probabilidade,
    "\n E(X)=",mean, "\n Var(X)=",var, "\n Assimetria=", skew, "\n Curtose=",kurt)

#FDA
fig, ax = plt.subplots(1, 1)
x = np.arange(nbinom.ppf(0.01, n, p),
              nbinom.ppf(0.99, n, p))
ax.plot(x, nbinom.pmf(x, n, p), 'bo', ms=8, label='nbinom pmf')
ax.vlines(x, 0, nbinom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
rv = nbinom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
        label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```

### Distribuição Hipergeométrica

Também é relacionado ao experimento de Bernoulli, porém a probabilidade de sucesso não é constante (sem reposição). A distribuição hipergeométrica descreve o número de sucessos na amostra de $n$ elementos, extraída de uma população finita sem reposição. Logo, descreve a probabilidade de que em uma amostra com $n$ elementos distintos extraídos da população sem reposição. \\

Consideremos uma população com $N$ elementos, dos quais $M$ possuem determinado atributo. A distribuição hipergeométrica descreve a probabilidade de que, em uma amostra com $n$ elementos distintos extraídos aleatoriamente da população sem reposição, exatamente $k$ possuem tal atributo ($k$ sucessos e $(n-k)$ fracassos). \\

Seja X uma variável aleatória, segue distribuição hipergeométrica com parâmetros N,M,n denotada por X~Hip(N,M,n), e sua FDP é dada por: \\

$P(X=k)=\frac{\left(\begin{matrix}M\\k\\\end{matrix}\right)*\left(\begin{matrix}N-M\\n-k\\\end{matrix}\right)}{\left(\begin{matrix}N\\n\\\end{matrix}\right)},\ \ \ 0\le k\le mín(M,n)$


```python
from scipy.stats import hypergeom
N=7
M=15
n=2
mean, var, skew, kurt = hypergeom.stats(M, n, N, moments='mvsk')
print(f"E(X)=",mean, "\n Var(X)=",var, "\n Assimetria=", skew, "\n Curtose=",kurt)
```

#### Exemplo Distribuição Hipergeométrica

Uma urna contém 15 bolas, das quais 5 delas são vermelhas. São escolhidas 7 bolas ao acaso, sem reposição. \\
Determine: \\
a) A probabilidade de que exatamente duas bolas vermelhas sejam sorteadas. \\
b) A probabilidade de que pelo menos duas bolas vermelhas sejam sorteadas. \\
e) O número esperado de bolas vermelhas sorteadas. \\
d) A variância do número de bolas vermelhas sorteadas. \\

```python
N=7 # Amostra
M=15 # População
n=5 # População com a característica

ex_a=hypergeom.pmf(2,15,5,7)
ex_b=(1-((hypergeom.pmf(0,15,5,7)+hypergeom.pmf(1,15,5,7))))
mean, var, skew, kurt = hypergeom.stats(M, n, N, moments='mvsk')

print("a=",ex_a,
      "\n b=", ex_b,
      "\n c=" , mean,
      "\n d=", var)

#FDA
rv = hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_dogs = rv.pmf(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_dogs, 'bo')
ax.vlines(x, 0, pmf_dogs, lw=2)
ax.set_xlabel('bolas vermelhas no grupo de bolas escolhida')
ax.set_ylabel('hypergeom PMF')
plt.show()
```

### Distribuição Poisson

Consiste em uma distribuição útil para registrar ocorrência de eventos Raros, com probabilidade de sucesso muito pequena $(p \rightarrow0)$, em determinada exposição (tempo, espaço). O modelo fornece a probabilidade do número de sucessos em determinado intervalo contínuo. \\

Por exemplo, quantidade de clientes que chegam na loja por unidade de tempo. Logo, a medida de exposição (tempo, área, espaço) é contínuo, porém a variável aleatória (número de ocorrências) é discreto. \\

Ela segue as seguintes hipóteses: \\
(i) Eventos definidos em intervalos não sobrepostos são independentes; \\
(ii) Em intervalos de mesmo comprimento, as probabilidades de ocorrência de um mesmo número de sucessos são iguais; \\
(iii) Em intervalos muito pequenos, a probabilidade de ocorrência de mais de um sucesso é desprezível; \\
(iv) Em intervalos muito pequenos, a probabilidade de um sucesso é proporcional ao comprimento do intervalo. 

Sua FDP é dada por $P\left(X=k\right)=\frac{e^{-\lambda}*\lambda^k}{k!},\ \ \ k=0,1,2,...$; \\
em que: \\
$e$= Logarítmo neperiano; \\
$\lambda$= taxa média estimada de ocorrência do evento de interesse. \\

Nessa distribuição a média é igual à variância:
$E(X)=Var(X)=\lambda$

```python
from scipy.stats import poisson
A = 0.6
mean, var, skew, kurt = poisson.stats(A, moments='mvsk')
print(f"E(X)=",mean, "\n Var(X)=",var, "\n Assimetria=", skew, "\n Curtose=",kurt)
```

#### Exemplo Poisson

Suponha que o número de clientes que chegam a um banco siga uma distribuição Poisson.Verifica-se que, em média, chegam 12 clientes por minuto. Calcule: \\
a) probabilidade de chegada de 10 clientes no próximo minuto; \\
b) probabilidade de chegada de 40 clientes nos próximos 5 minutos; \\
c) média e variância de X. 

```python
A=12
k=10
ex_a=poisson.pmf(k, A)
ex_b= poisson.pmf(8,12)
ex_c=A

print("a=",ex_a,
      "\n b=",ex_b,
      "\n c=",ex_c)

#FDA
fig, ax = plt.subplots(1, 1)
x = np.arange(poisson.ppf(0.01, A),
              poisson.ppf(0.99, A))
ax.plot(x, poisson.pmf(x, A), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, A), colors='b', lw=5, alpha=0.5)
rv = poisson(A)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
        label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```
